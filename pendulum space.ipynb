```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 0.2
b = 4.9
delta_t = 0.99999
steps = 100000
```

```python
def pendulum(x, y, a, b):
    x_1 = y
    y_1 = -a*y -b*np.sin(x)
    return x_1, y_1
```

```python
xx = np.empty((steps + 1))
yy = np.empty((steps + 1))
```

```python
xx[0], yy[0] = 0.000001, 1.0
```

```python
for i in range(steps):
    x_dot, y_dot = pendulum(xx[i], yy[i], a=a, b=b)
    xx[i + 1] = xx[i] + x_dot*delta_t
    yy[i+ 1] = yy[i] + y_dot*delta_t
```

```python
plt.figure(figsize = (10,6))
plt.style.use('dark_background')
plt.plot(xx, yy, '^', color='white', alpha = 0.9, markersize = 0.5)
plt.axis('on')
plt.show()
plt.close()
```
