```python
import numpy as np
import math
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
N = 500
steps = 500
a = np.zeros((steps, N), dtype = int)
initial_config = np.random.choice([0,1,2], size = (steps, N))
# initial_config = np.zeros((steps, N))
# initial_config[0, N//2] = 1
a = initial_config.copy()
```

```python
#Define the rules
rule_600 = {6: 0,
            5: 2,
            4: 1,
            3: 1,
            2: 0,
            1: 2,
            0: 0

           }
#{0, 2, 1, 1, 0, 2, 0} 
```

```python
# Apply Rule 600 to rows
def apply_rule_600(rows):
    new_rows = np.zeros_like(rows)
    for i in range(1, len(rows) - 1):
        new_rows[i] = rule_600[(rows[i-1] + rows[i] + rows[i + 1])]
    return new_rows
```

```python
#Now we want to apply rule 600 to the rows of matrix a

for t in range(1, steps):
    a[t] = apply_rule_600(a[t-1])
```

```python
# Define colors for each base 3 digit
colors = ['white', 'green', 'brown']
cmap = ListedColormap(colors)
#cmap = plt.cm.colors.ListedColormap(['white', 'gray', 'black'])
```

```python


# Plot the initial state
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.imshow(initial_config, cmap=cmap)
plt.title("Initial State")
plt.axis("off")

# Plot the final state
plt.subplot(1, 2, 2)
plt.imshow(a, cmap=cmap)
plt.title("Final State")
plt.axis("off")

plt.tight_layout()
plt.show()


```
