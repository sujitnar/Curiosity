```python
import numpy as np
import matplotlib.pyplot as plt
import math
from matplotlib.colors import ListedColormap
```

```python
N = 1000
steps = 9000
a = np.zeros((steps, N), dtype = int)
#initial_config = np.random.choice([0,1,2], size = (steps, N))
initial_config = np.zeros((steps, N))
initial_config[0, N//2] = 1
a = initial_config.copy()
```

```python
#Define the rules
rule_ = {6: 2,
        5: 0,
        4: 1,
        3: 2,
        2: 0,
        1: 2,
        0: 0

        }
#{2, 0, 1, 2, 0, 2, 0} 
```

```python
# Apply Rule 1635 to rows
def apply_rule(rows):
    new_rows = np.zeros_like(rows)
    for i in range(1, len(rows) - 1):
        new_rows[i] = rule_[(rows[i-1] + rows[i] + rows[i + 1])]
    return new_rows
```

```python
for t in range(1, steps):
    a[t] = apply_rule(a[t-1])
```
