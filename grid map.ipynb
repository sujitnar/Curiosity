```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a = 0.1
b = 0.1
delta_t = 0.5
steps = 1000000
```

```python
def grid_map(x, y, a, b):
    x_1 =  a*np.cos(y)
    y_1 =  b*np.sin(x)
    return x_1, y_1
```

```python
xx = np.empty((steps + 1))
yy = np.empty((steps + 1))
```

```python
xx[0], yy[0] = (1.0,0.0) 
```

```python
for i in range(steps):
    x_dot, y_dot = grid_map(xx[i], yy[i], a=a, b=b)
    xx[i+1] = xx[i] + x_dot*delta_t
    yy[i+1] = yy[i] + y_dot*delta_t
    
```

```python
plt.figure(figsize = (10,6))
plt.style.use('dark_background')
plt.plot(xx, yy, '^', color='red', alpha = 0.8, markersize = 0.05)
plt.axis('on')
plt.xlim(-50,10)
plt.ylim(40, -10)
plt.show()
plt.close()
```
