```python
import numpy as np
import matplotlib.pyplot as plt

# Define the continued exponential function
def continued_exponential(z, n_terms):
    result = np.exp(z)
    for i in range(1, n_terms-1):
        result = result*np.exp(result)
    return result

# Generate grid of complex numbers
x = np.linspace(-2, 2, 5000)
y = np.linspace(-2, 2, 5000)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Compute the continued exponential function for the grid of complex numbers
n_terms = 30
# Number of terms in the series
CE = continued_exponential(Z, n_terms)

# Plot the continued exponential function using domain coloring
plt.figure(figsize=(10, 8))
plt.imshow(np.angle(CE), extent=[-2, 2, -2, 2], cmap='viridis', origin='lower')
plt.colorbar(label='Phase (radians)')
plt.title('Continued Exponential Function using Domain Coloring')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```
