```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

n=10000
x = 1e-5 * np.ones(n)
iterations = 1000
last = 100
lam = np.linspace(2.5, 4.0, n)
def logistic(lam, x):
    return lam * x * (1 - x)
fig, ax1 = plt.subplots(figsize=(8, 9),
                               sharex=True)
for i in range(iterations):
    x = logistic(lam, x)
    # We compute the partial sum of the
    # Lyapunov exponent.
    
    # We display the bifurcation diagram.
    if i >= (iterations - last):
        ax1.plot(lam, x, ',k', alpha=.25)
ax1.set_xlim(2.5, 4)
ax1.set_title("Bifurcation diagram")
# n = 1000
# x = 0.5*np.ones(n)
# c = 0.4*np.ones(n)
# fig, ax1 = plt.subplots(figsize=(8, 9),
#                                 sharex=True)
# def logistic(c,x):
#     return x**2 + c
# for i in range(1000):
#     x= logistic(c,x)
#     if i>=(iterations-last):
#          ax1.plot(c,x, '.m',alpha=0.25)

   


    
```
