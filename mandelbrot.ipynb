```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#define size of grid and maximum number of iterations
size = 1000
max_iter = 100
```

```python
def mandelbrot(c):
    z= 0
    for i in range(max_iter):
        z = z**2 + c
        if abs(z) > 2:
            return i
    return max_iter
```

```python
#create a grid for complex numbers
x = np.linspace(-2,2,size)
y = np.linspace(-2,2,size)
X, Y = np.meshgrid(x,y)
Z = X + 1j*Y
```

```python
mandelbrot_set = np.zeros((size,size))
```

```python
for i in range(size):
    for j in range(size):
        mandelbrot_set[i,j] = mandelbrot(Z[i,j])
```

```python
plt.figure(figsize=(10, 10))
plt.imshow(mandelbrot_set, extent=[-2, 2, -2, 2], cmap='viridis', origin='lower')
plt.colorbar(label='Iterations to escape')
plt.title('Mandelbrot Set')
plt.xlabel('Real')
plt.ylabel('Imaginary')
plt.show()
```
